### INFORME DE PEDIDOS Y VENTAS

### Este dashboard contiene la necesaria informacin para analizar los pedidos y las ventas realizadas por un emprendimiento.

### Aqu se encuentran importantes indicadores tales como:
* #### Ventas del ao actual y el anterior
* #### Porcentaje de crecimiento
* #### Unidades vendidas.

### Adems podemos ver las ventas por productos, tipo de cliente, un resumen de los pedidos; por departamento y por mes.

### Todo ello, administrado por filtros por ao, tipo de cliente y departamento.
