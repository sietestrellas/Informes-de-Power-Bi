### INFORME DE PEDIDOS Y VENTAS

### Este dashboard contiene la necesaria información para analizar los pedidos y las ventas realizadas por un emprendimiento.

### Aquí se encuentran importantes indicadores tales como:
* #### Ventas del año actual y el anterior
* #### Porcentaje de crecimiento
* #### Unidades vendidas.

### Además podemos ver las ventas por productos, tipo de cliente, un resumen de los pedidos; por departamento y por mes.

### Todo ello, administrado por filtros por año, tipo de cliente y departamento.
