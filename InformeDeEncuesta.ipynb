### Informe Permanente de hogares correspondiente al año 2019

#### Este trabajo se desarrolló basado en los datos abiertos de la Provincia de Buenos Aires. 
#### Luego de la recopilación, el proceso de ETL(extracción, transformación y limpieza de datos), se procedió al análisis y la identificacion de métricas importantes como:
* ##### Total de encuestados
* ##### Porcentaje de mujeres y de hombres
* ##### Promedio de edad

#### Las visualizaciones detallan: 
* ##### Ingresos per cápita por sexo
* ##### Total de encuestados por afiliación Social
* ##### El máximo nivel educativo tanto en mujeres como en hombres
* ##### Comparativa de ingresos por situacion conyugal
* ##### Mapa donde se perciben colores como el rojo, amarillo y verde, especificándo el nivel de ingresos por comuna.
