### INFORME DE LOGISTICA

#### Este informe se ha diseñado con la intención de analizar la situación general del departamento de logistica de una empresa.
#### Encontrando los KPI:

* ##### Total facturado
* ##### Cantidad de facturas
* ##### Cantidad enviada
* ##### Flete en Pesos
* ##### Peso bruto

#### Un dato importante es la identificación de la métrica de suministro como es el OTIF (On time In full), la cual tiene como propósito medir la capacidad de una empresa para cumplir con los pedidos de los clientes.
#### Visualizando este indicador por año y mes, por ciudad, además de comparararlo con el porcentaje de objetivo.
#### En cuanto a los transportistas se realizaron gráficos basados la ocurrencia y la entrega a tiempo de los pedidos.
#### El filtro hace referencia a los últimos tres años de la empresa.
